```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
# Multiple vectorsin a single graph
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,7, scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,8,-9, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-2,-9, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
# Addition of two vectors
vector1 = np.array([4,5])
vector2 = np.array([-3,7])
vector3 = vector1 + vector2
print(vector3)
```

```python
plt.quiver(0,0,vector1[0],vector1[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(vector1[0],vector1[1],vector2[0],vector2[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,vector3[0],vector3[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-1,12)
plt.ylim(-1,13)
plt.show()
```

```python
# Another way
vector1 = np.array([0,0,4,5])
vector2 = np.array([0,0,-3,7])
vector3 = vector1 + vector2
print(vector3)
```

```python
plt.quiver(vector1[0],vector1[1],vector1[2],vector1[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(vector2[0],vector2[1],vector2[2],vector2[3], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(vector3[0],vector3[1],vector3[2],vector3[3], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-6,6)
plt.ylim(-1,13)
plt.show()
```

```python
# Subtraction of two vectors
vector1 = np.array([4,5])
vector2 = np.array([-3,7])
vector3 = vector1 - vector2
print(vector3)
```

```python
plt.quiver(0,0,vector1[0],vector1[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(vector1[0],vector1[1],vector2[0],vector2[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(0,0,vector3[0],vector3[1], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-1,8)
plt.ylim(-5,13)
```

```python
# Another way
vector1 = np.array([0,0,4,5])
vector2 = np.array([0,0,-3,7])
vector3 = vector1 - vector2
print(vector3)
```

```python
plt.quiver(vector1[0],vector1[1],vector1[2],vector1[3], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(vector2[0],vector2[1],vector2[2],vector2[3], scale_units='xy', angles='xy', scale=1, color='g')
plt.quiver(vector3[0],vector3[1],vector3[2],vector3[3], scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-4,8)
plt.ylim(-5,8)
plt.show()
```

```python
# Multiplication of a vector by a scalar
vector1 = np.array([4,5])
vector2 = 3 * vector1
print(vector2)
```

```python
plt.quiver(0,0,vector1[0],vector1[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,vector2[0],vector2[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-1,15)
plt.ylim(-1,20)
plt.show()
```

```python
# Multiplication of a vector by a scalar
vector1 = np.array([4,5])
vector2 = -0.5 * vector1
print(vector2) 
```

```python
plt.quiver(0,0,vector1[0],vector1[1], scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,vector2[0],vector2[1], scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-3,5)
plt.ylim(-5,6)
plt.show()
```
