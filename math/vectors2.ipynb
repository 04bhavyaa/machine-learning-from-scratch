Dot Product

```python
# dot product
v1 = [1, 2, 3]
v2 = [4, 5, 6]

dp = sum([v1[i] * v2[i] for i in range(len(v1))])
print(dp)
```

```python
import numpy as np
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(v1.dot(v2))
```

```python
v1 = np.array([1, 2])
v2 = np.array([4, 5, 6])

try:
    print(v1.dot(v2))
except Exception as e:
    print(e)
    print("Dot product is not defined for vectors of different length")
```

```python
v1 = np.array([1, 2])
v2 = np.array([4, 5])

v = np.dot(v1, v2)
v
```

Cross Product

```python
# Cross product
v1 = [1, 2, 3]
v2 = [4, 5, 6]

v = np.cross(v1, v2)
v
```

Projection of Vector

```python
# Projection
v1 = np.array([1, 2])
v2 = np.array([3, 4])
# Magnitude of v2
mag_v2 = np.sqrt(sum(i**2 for i in v2))
v= (np.dot(v1, v2) / mag_v2**2)*v2
v
```

```python
# plot cross product of two vectors in 2d
import matplotlib.pyplot as plt

v1 = np.array([1, 2])
v2 = np.array([3, 4])
mag_v2 = np.sqrt(sum(i**2 for i in v2))
v = (np.dot(v1, v2) / mag_v2**2)*v2
print(v)

fig = plt.figure()
fig.suptitle('Cross product of two vectors')
plt.plot([0, v1[0]], [0, v1[1]], 'r', label='v1')
plt.plot([0, v2[0]], [0, v2[1]], 'g', label='v2')
plt.plot([0, v[0]], [0, v[1]], 'b', label='v1 x v2')
plt.grid()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.show()
```
